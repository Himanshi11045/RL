```python
import numpy as np
```

```python
np.random.seed(1)
```

```python
x = np.random.randint(0,50,1000)
print(x)
```

```python
y =x+ np.random.normal(0,10,1000)
print(y)
```

```python
np.corrcoef(x,y)
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.style.use('ggplot')
plt.scatter(x,y)
plt.show()
```

```python
y =x- np.random.normal(0,10,1000)
print(y)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
y = 100 -x +np.random.normal(0,5,1000)
np.corrcoef(x,y)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
z = np.random.normal(0 , 30, 1000)
print(z)
```

```python
y = z+np.random.normal(0,10,1000)
print(y)
```

```python
np.corrcoef(z,y)
```

```python
x = z+np.random.normal(0,50,1000)
print(x)
```

```python
import pandas as pd
df = pd.DataFrame({'a': np.random.randint(0, 50,1000)})
df['b'] = df['a'] + np.random.normal(0,10,1000)
df['c'] = 100-df['a'] + np.random.normal(0,5,1000)
df['d'] = np.random.randint(0,50,1000)
```

```python
pd.plotting.scatter_matrix(df,figsize = (6,6))
plt.show()
```

```python
plt.matshow(df.corr())
plt.xticks(range(len(df.columns)),df.columns)
plt.yticks(range(len(df.columns)), df.columns)
plt.colorbar()
plt.show
```
